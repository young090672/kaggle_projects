```python
corr = train.corr(numeric_only=True)['SalePrice'].sort_values()
corr
```

```python
low_corr_cols = corr[abs(corr) < 0.1].index.tolist()

print(low_corr_cols)
train = train.drop(columns=[col for col in low_corr_cols if col != 'Id'])
test  = test.drop(columns=[col for col in low_corr_cols if col != 'Id'])
```

```python
cat_cols = train.select_dtypes(include=['object']).columns.tolist()
drop_cols=[]
for col in cat_cols:
  top_ratio = train[col].value_counts(normalize=True, dropna=False).iloc[0]
  if top_ratio > 0.95:
    drop_cols.append(col)
    print(f"{col}:{top_ratio:.3f} drop 대상")
print("\n드랍할 컬럼:",drop_cols)
train = train.drop(columns=drop_cols)
test = test.drop(columns=drop_cols)
```

```python
train = train.drop(columns='Alley')
test = test.drop(columns='Alley')
```

```python
none_cols = [
    'GarageType','GarageFinish','GarageQual','GarageCond',
    'BsmtQual','BsmtCond','BsmtExposure','BsmtFinType1','BsmtFinType2',
    'PoolQC','MiscFeature','Alley','Fence','FireplaceQu'
]

for col in none_cols:
    if col in train.columns:
        train[col] = train[col].fillna("None")
        test[col]  = test[col].fillna("None")
```

```python
cat_cols = train.select_dtypes(include=['object']).columns.tolist()

for col in cat_cols:
    if train[col].isna().sum() > 0:   # NA 있는 경우만
        train[col] = train[col].fillna("None")
        test[col]  = test[col].fillna("None")
```

```python
train['LotFrontage'] = train.groupby('Neighborhood')['LotFrontage'].transform(
    lambda x: x.fillna(x.median())
)
test['LotFrontage'] = test.groupby('Neighborhood')['LotFrontage'].transform(
    lambda x: x.fillna(x.median())
)
```

```python
train['MasVnrArea'] = train['MasVnrArea'].fillna(0)
test['MasVnrArea'] = test['MasVnrArea'].fillna(0)
```

```python
# 차고가 없는 경우: 0으로 채움
train.loc[train['GarageType'] == "None", 'GarageYrBlt'] = 0
test.loc[test['GarageType'] == "None", 'GarageYrBlt'] = 0

# 차고가 있는데 NA이면 건축연도(YearBuilt)로 채움
train['GarageYrBlt'] = train['GarageYrBlt'].fillna(train['YearBuilt'])
test['GarageYrBlt']  = test['GarageYrBlt'].fillna(test['YearBuilt'])
```

```python
train.columns
```

```python
# 1) 테스트용 PassengerId 저장 (제출용)
test_id = test['Id']

# 2) 학습(X)에서는 PassengerId 제거
train = train.drop('Id', axis=1)
```

```python
test = test.drop('Id', axis=1)
```

```python
cat_cols = train.select_dtypes(include=['object']).columns.tolist()
print(cat_cols)
```

```python
# 1) train + test 모두 One-Hot Encoding
train = pd.get_dummies(train, columns=cat_cols, drop_first=True)
test  = pd.get_dummies(test,  columns=cat_cols, drop_first=True)

# 2) train/test 컬럼을 동일하게 맞추기
train, test = train.align(test, join='left', axis=1)

# 3) test에서 새로 생긴 NA는 0으로 채움
test = test.fillna(0)
```

```python
# Target 설정
target = ['SalePrice']

# 데이터 분리
X = train.drop(target, axis=1)
y = train.loc[:, target]
```

```python
# 모듈 불러오기
from sklearn.model_selection import train_test_split

# 7:3으로 분리
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size = 0.2, random_state=1)
```

```python
from sklearn.ensemble import RandomForestRegressor
from lightgbm import LGBMRegressor
from xgboost import XGBRegressor

from sklearn.metrics import *
```

```python
model1 = RandomForestRegressor(n_estimators=300,
                                max_depth=15,
                                random_state=42,
                                n_jobs=1)
```

```python
model1.fit(X_train, y_train)
```

```python
y_pred = model1.predict(X_val)
```

```python
print(r2_score(y_val,y_pred))
print(mean_absolute_error(y_val,y_pred))
print(root_mean_squared_error(y_val,y_pred))
```

```python
model2 = LGBMRegressor(
    n_estimators=500,
    learning_rate=0.05,
    random_state=42
)
```

```python
model2.fit(X_train, y_train)
```

```python
y_pred2 = model2.predict(X_val)
```

```python
print(r2_score(y_val,y_pred2))
print(mean_absolute_error(y_val,y_pred2))
print(root_mean_squared_error(y_val,y_pred2))
```

```python
model3 = XGBRegressor(
    n_estimators=4000,
    learning_rate=0.03,
    max_depth=4,
    subsample=0.7,
    colsample_bytree=0.7,
    reg_alpha=0.1,
    reg_lambda=1,
    random_state=42,
    tree_method='hist'
)
```

```python
y_pred3 = model3.predict(X_val)
```

```python
best_model = RandomForestRegressor(n_estimators=300,
                                max_depth=15,
                                random_state=42,
                                n_jobs=1)

best_model.fit(X, y)
```
